• define a function missingdf(df) with input argument df which could print out all the column in dataframe df and also the missing value rate for each column. For example, with total 1000 rows, if column1 has missing value in 200 rows and its missing value rate will be 0.2 or 20%. The result of the function missingdf(df) will print the new dataframe which has two columns: the column_name and the percent_missing

**Answer 1.2**

```python
#Defining the Function
def missingdf(df):
  missing_value_rate = df.isnull().sum()/ len(df)
  #New dataframe with new columns
  missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': missing_value_rate},index=None)
  return missing_value_df

print(missingdf(df1).to_string(index=False))
```

I defined a fucntion missingdf(df), which can find the percent of missing value by dividing the empty values in each rows by the total number of rows in the dataframe. Then I created a new dataframe that has to columns column_name and the percent_missing.

```python
#Grpuping columns to form a basket
grp_df = cust_tran_df.groupby(['customer_id', 'basket_id'])
grp_df['commodity'].value_counts() #returns commodity in each basket
```

```python
#Number of transactions based on basket level
trans_count = cust_tran_df.groupby(['customer_id', 'basket_id']).agg({'commodity': ['count']})
print('The number of transactions based on basket level is ' + str((trans_count).shape[0]))#no. of rows is the total no. of transactions
avg_basket_size = trans_count.mean()
avg_basket_size
```

```python
#Most popular product commodity
cust_tran_df['commodity'].value_counts()
```

```python
#average basket total price for each customer
average_basket_total_price = cust_tran_df.groupby(['customer_id', 'basket_id']).agg({'price': 'mean'})
average_basket_total_price
```

```python
#Transforming the data into basket
copied_df = cust_tran_df.copy(deep=True)
transaction_product = pd.crosstab(copied_df.basket_id, copied_df.commodity).astype('bool') #Transforms dataset and enters boolean values
transaction_product.columns.name='commodity'
transaction_product.index.name='basket_id'
transaction_product.head()
```

```python
customer_product_freq = pd.crosstab(copied_df.customer_id, copied_df.commodity).astype('int') #Transforms dataset and enters integer values
customer_product_freq.columns.name='commodity'
customer_product_freq.index.name='customer_id'
customer_product_freq.tail(10)
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler

#Normalizing the data
scaler = StandardScaler()
customer_product_freq_scaled = scaler.fit_transform(customer_product_freq)

#Implementing KNN to find the top 5 similar customers for each customer
knn = NearestNeighbors(n_neighbors=6)  # n_neighbors is set to 6 to find the top 5 neighbors excluding the customer itself
knn.fit(customer_product_freq_scaled)

#Finding the top 5 nearest neighbors for each customer
distances, indices = knn.kneighbors(customer_product_freq_scaled)

#Extracting the top 5 similar customers for each customer
similar_customers = {}
for i, customer_id in enumerate(customer_product_freq.index):
    # Skipping the first neighbor as it is the customer itself
    similar_customers[customer_id] = customer_product_freq.index[indices[i][1:6]].tolist()

#Displaying similar customers for the first few customers as an example
example_customers = list(similar_customers.keys())[:5]
for customer_id in example_customers:
    print(f"Customer {customer_id} similar customers: {similar_customers[customer_id]}")
```
